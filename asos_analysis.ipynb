```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('products_asos.csv', on_bad_lines='skip')

#forcing every error to none(NA)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
#if price is not a number, whole column will be dropped
df = df.dropna(subset=['price'])

#printing number of rows in the data
print(f"Data Loaded: {len(df)} rows")

df.head()
```

```python
#cleaning up the description
#making sure that description column is string

df['description'] = df['description'].astype(str)
 #text splitting in description

def get_brand(text):
  if 'by ' in text:
    try:
      return text.split('by ')[1].split(' ')[0]
    except:
      return "Unknown"
  return "Unknown"

df['brand_raw'] = df['description'].apply(get_brand)
```

```python
df.head(3)
```

```python
brand_map = {
    'New': 'New Look',
    'River': 'River Island',
    'Miss': 'Miss Selfridge',
    'TopshopWelcome': 'Topshop',
    "Vero": "Vero Moda",
    "The": "The North Face",
    "Urban": "Urban Revivo",
    "In": "In The Style",
    "Noisy": "Noisy May",
    "Reclaimed": "Reclaimed Vintage",
    "Daisy": "Daisy Street",
    "Pretty": "PrettyLittleThing",
    "Extro": "Extro & Vert",
}
df['brand_raw'] = df['description'].apply(get_brand)
df['Brand'] = df['brand_raw'].map(brand_map).fillna(df['brand_raw'])

#brand count
brand_counts = df['Brand'].value_counts()
valid_brands = brand_counts[brand_counts > 5].index
df_clean = df[df['Brand'].isin(valid_brands)].copy()

print(df_clean['Brand'].value_counts().head(5))
```

```python
#cleaning the size column

#1.Function to analyze stockouts
def calculate_phantom_revenue(size_str):
  if not isinstance(size_str, str):
    return 0, 0.0

  #split"UK 6, UK 8 - Out of Stock" into list
  sizes = size_str.split(',')
  total_sizes = len(sizes)

  #count how many items are out of stock
  out_of_stock_count = size_str.count('Out of stock')

  #Calculate Rate(0.0 to 1.0)
  rate = out_of_stock_count / total_sizes if total_sizes > 0 else 0.0

  return out_of_stock_count, rate

metrics = df_clean['size'].apply(lambda x: calculate_phantom_revenue(x))
df_clean['Stockout_Count'] = [x[0] for x in metrics]
df_clean['Stockout_Rate'] = [x[1] for x in metrics]

df_clean['Lost_Revenue'] = df_clean['Stockout_Count'] * df_clean['price']
#
cols = ['Brand', 'name', 'price', 'Stockout_Count', 'Lost_Revenue']
print(df_clean.sort_values(by='Lost_Revenue', ascending=False).head(5)[cols])
```

```python
#summarizing everything done

brand_strategy = df_clean.groupby('Brand').agg({
    'price': 'mean',
    'Stockout_Rate': 'mean',
    'Lost_Revenue': 'sum',
    'name': 'count'
}).reset_index()

brand_strategy = brand_strategy[brand_strategy['name'] > 10]

plt.figure(figsize=(12, 8))
sns.scatterplot(
    data = brand_strategy,
    x = 'price',
    y = 'Stockout_Rate',
    size = 'Lost_Revenue',
    sizes = (50,500),
    alpha = 0.7,
    palette = 'viridis'
)

winners = brand_strategy[
    (brand_strategy['price']>40)&
    (brand_strategy['Stockout_Rate']>0.4)
]

for i in range(len(winners)):
  plt.text(
      winners.iloc[i]['price']+1,
      winners.iloc[i]['Stockout_Rate'],
      winners.iloc[i]['Brand']
  )

plt.title('Brand Strategy Analysis')
plt.xlabel('Average Price')
plt.ylabel('Stockout Rate')
plt.axvline(x = 40, color = 'red', linestyle = '--')
plt.axhline(y = 0.4, color = 'red', linestyle = '--')
plt.show()

#------------------##observations##----------------------------------------#
#most of inventory is clustered in low demand, low price so low growth, items not much useful for the website
#top right, high demand brands selling out faster than restocking rate
#even with high price, customers are willing to buy from thesebrands
#top left, low price, high demand, items that people constantly need and constantly buy
#lower right quadrant, things, website might think to liquidate
# recommended to look at top right, useful for the website
```
